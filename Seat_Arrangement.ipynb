```python
import pandas as pd
n=int(input('Number of Students in a class : '))
t=[i for i in range(1,n+1)]
e=int(input('Value of e: '))
o=int(input('Value of o: '))
me=[x%n for x in t]
# print(me)
ne=[p+e if p%2==0 else p+o for p in me]
# print(ne)
te=[x%n for x in ne]
# print(te)
tee=[n if x==0 else x for x in te]
et=[x for x in tee if tee.count(x)>1]
et=list(set(et))
gt=[tee.index(x) for x in et]
# print(gt)
ki=[x for x in t if x not in tee]
print('Initial:',t)
print('After:',ne)
# Relacing the repeated elements in two lists
for i in range(0,len(et)):
    tee[tee.index(et[i])]=ki[i]
print('Final:',tee)    
if set(t)==set(tee):
    print('All Good')
    df1=pd.DataFrame(t,columns=['Current Seat number'])
    df2=pd.DataFrame(tee,columns=['Next Seat number'])
    df1["New seat number"]=df2['Next Seat number']
    df1.index=df1.index+1
#     print(df1)
else:
    print('Something is wrong')
df1
```
